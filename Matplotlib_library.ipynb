#**BAR PLOT**

```python
sub=['Python','c','c++','OS']
marks=[95,80,70,90]
col=['y','g','b','r']
plt.bar(sub,marks,width=0.5,color=col,edgecolor='purple',linewidth=10)
plt.title("Result",fontsize=40)
plt.xlabel("Marks Obtained",fontsize=20)
plt.ylabel("Subject",fontsize=20)
plt.show()
```

```python
sub=['Python','c','c++','OS']
marks=[95,80,70,90]
col=['y','g','b','r']
plt.bar(sub,marks,width=0.5,color=col,edgecolor='purple',linewidth=5,linestyle=':',alpha=0.6)
plt.title("Result",fontsize=40)
plt.xlabel("Marks Obtained",fontsize=20)
plt.ylabel("Subject",fontsize=20)
plt.show()
```

```python
sub=['Python','c','c++','OS']
marks1=[95,80,70,90]
marks2=[20,35,40,60]
col=['y','g','b','r']
plt.bar(sub,marks1,width=0.5,color='r',label='popularity1')
plt.bar(sub,marks2,width=0.5,color='g',label='popularity2')
plt.title("Result",fontsize=40)
plt.xlabel("Marks Obtained",fontsize=20)
plt.ylabel("Subject",fontsize=20)
plt.legend()
plt.show()
```

```python
import numpy as np
sub=['Python','c','c++','OS']
marks1=[95,80,70,90]
marks2=[20,35,40,60]

width=0.4
p=np.arange(len(sub))
p1=[j+width for j in p]

plt.bar(p,marks1,width,color='r',label='popularity1')
plt.bar(p1,marks2,width,color='g',label='popularity2')

plt.title("Result",fontsize=40)
plt.xlabel("Marks Obtained",fontsize=20)
plt.ylabel("Subject",fontsize=20)

plt.xticks(p+width/2,sub,rotation=10)
plt.legend()
plt.show()
```

#**SCATTER PLOT**

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color='r')
plt.title("Scatter plot",fontsize=15)
plt.xlabel("day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
col=['r','g','b','r','g','b','r']
size=[100,200,400,500,600,700,500]
plt.scatter(day,no,color=col,s=size,alpha=0.9,marker='*',edgecolors='black',linewidths=2)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
col=[10,33,40,50,70,40,90]
size=[100,200,400,500,600,700,500]
plt.scatter(day,no,c=col,s=size,cmap='viridis')
t=plt.colorbar()
t.set_label('colorbar',fontsize=15)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("day",fontsize=10)
plt.ylabel("NO.",fontsize=10)
plt.show()
```

#**PIECHART**

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['C','C++','Java','Python']
ex=[0.4,0.0,0.0,0.0]
col=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=col,autopct='%0.1f%%',shadow=True,radius=1.5,labeldistance=1.2,textprops={'fontsize':15},wedgeprops={'linewidth':4})
#startangle=0,90,180,270
#rotatelabels=True
#counterclock=False
#'edgecolor':'m'

plt.show()
```

```python
#dot pie chart
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=['C','C++','Java','Python']


plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=0.5)

plt.show()
```
